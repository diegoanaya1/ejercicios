```python
pl=df["Posesion Real Madrid Local"]=round((dfmadridlocal["Home Team Possession %"]).sum()/133,2)
pv=df["Posesion Real Madrid Visita"]=round((dfmadridvisita["Away Team Possession %"]).sum()/133,2)


```

```python
#Creamos la grafica mediante los datos de la posesion

pl=round((dfmadridlocal["Home Team Possession %"]).sum()/133,2)
pv=round((dfmadridvisita["Away Team Possession %"]).sum()/133,2)
# put them into a list called proportions
proportions = [59.93, 40.47]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['Posesion Real Madrid en casa', 'Otros equipos'],
    
    # with no shadows
    shadow = True,
    
    # with colors
    colors = ['blue','yellow'],
    
    # with one slide exploded out
    explode = (0.15 , 0.5),
    
    # with the start angle at 90%
    startangle = 200,
    
    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Posesion Promedio del Real Madrid en casa")


# View the plot
plt.tight_layout()
plt.show()

# put them into a list called proportions
proportions = [57.58, 42.42]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,
    
    # with the labels being officer names
    labels = ['Posesion Real Madrid de visitante', 'Otros equipos'],
    
    # with no shadows
    shadow = True,
    
    # with colors
    colors = ['red','yellow'],
    
    # with one slide exploded out
    explode = (0.15 , 0.5),
    
    # with the start angle at 90%
    startangle = 200,
    
    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Posesion Promedio del Real Madrid de visitante")

# View the plot
plt.tight_layout()
plt.show()
```

```python
dfpmlocal=df["Rating Real Madrid Local"]=round((dfmadridlocal["Home Team Rating"]).sum()/133,2)
dfpmvisita=df["Rating Real Madrid Visita"]=round((dfmadridvisita["Away Team Rating"]).sum()/133,2)
dfpmlocal
```

```python
dfpmlocal=df["Rating Real Madrid Local"]=round((dfmadridlocal["Home Team Rating"]).sum()/133,2)
dfpmvisita=df["Rating Real Madrid Visita"]=round((dfmadridvisita["Away Team Rating"]).sum()/133,2)
local=(dfpmlocal)
visita=(dfpmvisita)
grafica= ["local", "visita"] # Items de la grafica
rating = [local,visita] # Valores de los items
colors= ['green','blue'] # Colores de los items
plt.bar(grafica,rating,color=colors) # asignar items, valores y colores a la grafica
plt.title('Rating') # colocar un titulo a la grafica
plt.xlabel('Localidad') # Poner un titulo o etiqueta a el Eje X
plt.ylabel('Nivel deRating') # Poner un titulo o etiqueta a el Eje Y
plt.show() # Mostrar la grafica
```

```python
dfexcvisita=df["Media de emocion en los partidos del Real Madrid de visita"]=round((dfmadridlocal["Match Excitement"]).sum()/133,2)
dfexclocal=df["Media de emocion en los partidos del Real Madrid en casa"]=round((dfmadridvisita["Match Excitement"]).sum()/133,2)
locald=(dfexclocal)
visitad=(dfexcvisita)
grafica=["local","visita"]
valor=[locald,visitad]
colors=["red","blue"]
plt.bar(grafica,valor,color=colors)
plt.title('Nivel de emocion de los partidos') # colocar un titulo a la grafica
plt.xlabel('') # Poner un titulo o etiqueta a el Eje X
plt.ylabel('Promedio Nivel de emocion') # Poner un titulo o etiqueta a el Eje Y

plt.show() # Mostrar la grafica
```
